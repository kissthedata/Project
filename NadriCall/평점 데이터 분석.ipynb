# 평점 데이터 분석

```python
ratings = pd.read_csv("./추가 나드리콜 데이터/2. 고객 평점 데이터/driver_rating.csv")
ratings.drop(['Unnamed: 0'],axis=1, inplace=True)
```

# 전처리

```python
#to_datetime
ratings['날짜'] = pd.to_datetime(ratings['날짜'])

#관제탑 시스템이 도입된 21년 이후로 분석
ratings = ratings[ratings['날짜'] > '2022-01-01']
```
